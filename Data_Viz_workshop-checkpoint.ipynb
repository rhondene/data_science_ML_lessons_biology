## 1. Scatter plot

- Widely used for visualizing relationships between 2 (or even more) variables

```python
##how does sepal length compare to petal length?
plt.figure(figsize=(8,6))
seaborn.scatterplot(data=iris, x= 'sepal length',y='petal length')

#adjust x and y axes labels
plt.show()
```

```python
##how does sepal length compare to petal length?
plt.figure(figsize=(8,6))
seaborn.scatterplot(data=iris, x= 'sepal length',y='petal length', hue='species',palette='Set1') #color by species

plt.show()
```

```python
#adjust marker just in case graph printed in black and white
plt.figure(figsize=(8,6))
seaborn.scatterplot(data=iris, x= 'sepal length',y='petal length', 
                    hue='species',palette='Set1', style='species') #color by species
plt.show()
```

## 1b. Bubble plot

A bubble plot is a scatterplot where a third dimension is added: the value of an additional numeric variable is represented through the size of the dots. (source: data-to-viz.com).

- We will modify the scatterplot into a bubbleplot by adding petal width as the third dimension

```python
#all points have single color
plt.figure(figsize=(8,6))
seaborn.scatterplot(data=iris, x= 'sepal length',y='petal length', hue='species', size='petal width'
                     ) #color by species
#adjust legend
plt.legend(bbox_to_anchor=(1, 1)) #(horizontal direction, vertical alignment)
plt.show()
```

## 1c. Regression plot
- Seaborn make statistical plotting convenient
- Seaborn's <b>regplot()</b>  draws a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

Linear regression answers: how much of variable y can be explained by x? i.e. statistical relationship between x an y

```python
plt.figure(figsize=(8,6))
seaborn.regplot(data=iris, x='petal length', y='sepal length', scatter=True, color='black',
               scatter_kws={'color':'red'})
plt.show()
```

- By default, barplot() calculate the <i>mean</i> for each category. That is, the length of each bar is the mean value of that category. You can change to a different summary statistic using the "estimator" option
- By default, the error bars represent the 95% confidence interval but we can change to the standard deviation (measures variability)

```python
#visualise how sepal width varies between species
plt.figure(figsize=(8,6))
#make error bars reprsent standard deviation 
seaborn.barplot(data=iris, x='species',y='sepal width', ci='sd', edgecolor='black') 
plt.show()
```

##  4.  Histogram for univariate distributions

One of the first things we do with new data is to explore the distribution of the data and its statistics. 
- Histograms are used for visualizing the frequency distribution of a <i> continuous </i> variable (univariate) 
- Histograms display the frequency with which a variable takes on a particular value
- A histogram displays numerical data by grouping data into "bins" of equal width. Each bin is plotted as a bar whose height corresponds to how many data points are in that bin.

### Histogram and bins
- <b> Number of Bins </b> The choice of the number of bins is important as it controls the shape of the distribution (number of bars). In turn, the shape of the distribution can inform the conclusions we draw from data. 
- Most times seaborn and matplotlib does good enrough job of choosing the default number of bins BUT
- Always experiment with different bin sizes for a given data sample to get multiple perspectives or views on the same data.
- While there is no strict rule on how many bins to usejust avoid using too few or too many bins.

```python
# a single bin
plt.figure(figsize=(8,6))
seaborn.histplot(data=iris,x='petal length', bins=1)
plt.show()
```

##### clearly the distribution of petal length is not unform. Rather it varies accross species as we have seen above
- The wrong number of bins can mislead our inference.

```python
# play around with bins number 
fig, axes = plt.subplots(figsize=(12,4), nrows=1, ncols=3)  #this generates 3 [sub]plots (ncols=3) on the same figure
seaborn.histplot(data=iris,x='petal length', bins=5, ax=axes[0], color='#f5a142') #  #f5a142 is a hex color
seaborn.histplot(data=iris,x='petal length', bins=10, ax=axes[1], color='maroon')
seaborn.histplot(data=iris,x='petal length', bins=15, ax=axes[2], color='darkmagenta')

plt.tight_layout() #an alternative to plt.show(); tight_layout adds space between individual plot
```

### 4b.  Kernel Density Estimation (KDE plot) plot 
- A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analagous to a histogram, but more sophisticated. <b> No need to guess-timate bins!! </b> 
-  KDE fits a curve on the data that represents the estimated probability density function (PDF) for that dataset.
- KDE plot is useful for revealing the underlying shape of the distribution, and choosing the number of bins for a histogram

```python
#plot a kdeplot
plt.figure(figsize=(8,6))  
seaborn.kdeplot(data=iris,x='petal length')
plt.show()
```

##### notice the y-axis changed to 'Density' which may be hard to interpret non-stats folks
- This is why I prefer to plot a kde overlaying a histogram. 
- Recall that the y-axis on histogram is the frequency at which a variable (petal length) takes on a value. 
- The y-axis of a histogram is more interpretable for an audience than explaining what "Density" means
 more info on interpreting raw kdeplot https://jakevdp.github.io/PythonDataScienceHandbook/05.13-kernel-density-estimation.html

```python
#plot a kde plot using histplot but hiding the histogram so that only kdeplots are available
plt.figure(figsize=(8,6)) 
seaborn.histplot(data=iris,x='petal length',kde=True, alpha=0.0, edgecolor='white',
                 hue='species') #alpha controls the transparency of the histogram,
plt.title('KDE Plot')
plt.show()
```

```python
#  KDE plot often reveal more accurate approximation of the true distibtution vs. arbitrary bin size
plt.figure(figsize=(8,6))  
seaborn.histplot(data=iris,x='petal length', bins=1, kde=True)
plt.show()
```

```python
# regular boxplot
plt.figure(figsize=(8,6))
seaborn.boxplot(data=iris,x='species', y='petal length')
plt.show()
```
